```python
import pymc3 as pm
import numpy as np
import scipy as sci
import matplotlib.pyplot as plt
```

### Цель в нахождении оптимальных параметров распределений, которые гипотетически описывают распределение случайной величины(вид распределения задается аналитиком)
$$
Простейшая \space модель
\\
in1 \sim Norm(a_1,\theta_1 )
\\
in2 \sim Norm(a_2,\theta_2 )
\\
out \sim Norm(a_3,\theta_3)
\\
in1+in2=out
$$

```python
## Создадим генератор наблюдений
norm_in1=sci.stats.norm(loc=30,scale=10)
norm_in2=sci.stats.norm(loc=50,scale=5)
norm_out=sci.stats.norm(loc=80,scale=15)
size=1000 ##объем наблюдений
```

```python

with pm.Model() as model:
    #Параметры, которые после использования методов PyMC должны задавать распредения in1,in2,out с учетом
    #уравнений и наблюдений
    
    #Матожидания(их начальные параметры могут быть произвольными, алгоритм их все равно изменяет)
    mu_in1=pm.Normal("mu_in1")
    mu_in2=pm.Normal("mu_in2")
    mu_out=pm.Deterministic("mu_out",mu_in1+mu_in2)
    
    #Дисперсии
    sd_in1=pm.Uniform('sd_in1',1,15)
    sd_in2=pm.Uniform('sd_in2',1,10)
    sd_out=pm.Uniform('sd_out',1,25)
    
    #Те самые распределения, которые мы хотим "оптимизировать"   
    in1=pm.Normal("in1",mu=mu_in1,sd=sd_in1,observed=norm_in1.rvs(size))
    in2=pm.Normal("in2",mu=mu_in2,sd=sd_in2,observed=norm_in2.rvs(size))
    out=pm.Normal("out",mu=mu_out,sd=sd_out,observed=norm_out.rvs(size))
```

```python
with model:
    start=pm.find_MAP()
    print(start)
    step=pm.NUTS()
    trace=pm.sample(6000,step=step,tune=1500,chains=2)
    
```

Можно заметить насколько сильно изменилось распределение дисперсий после алгоритма.
PyMC изменяет распределения параметров

```python
pm.traceplot(trace)
```

Заметим, что pm.find_MAP() дало как раз значения mean для параметров(то есть не обязательно использовать каждый раз NUTS)

```python
data=pm.summary(trace)
data
```

### Графики апосториорных распределений случайных величин с учетом наблюдений и уравнений связи. После нахождения "оптимальных" параметров

### Попробуем с учетом построенных распределений оптимизировать вновь приходящие наблюдения

$$
\\
Функционалы  \space ошибок(при \space условиях \space уравнений \space связи)
\\
Стандартный \space функционал \space - квадратичное \space отклонение
\\
\min_{\bar{y} \in \mathbb{R}^n}  \sum_{i=1}^{n}(y_i-x_i)^2
\\
\text{Функционалы, которые предлагаем ввести мы, чтобы учесть распределения величин.}
\\
\text{Т.к. мы минимизируем функцию, то вероятности в функционале должны быть обратны значению функционала}
\\
\min_{\bar{y} \in \mathbb{R}^n}  \sum_{i=1}^{n}((y_i-x_i)^2*(1-\mathbb{P}_i([x_i-\varepsilon_i,x_i+\varepsilon_i]))
\\
\text{Значения эпсилон будем выбирать в зависимости от распредения.}
\\
\text{Возможно есть более качественные методы оценки вероятности измерений. Мы еще думаем над этим.}
$$

```python
def least_squares(x,y):  
    #Cтандартный функционал ошибки 
    return (x[0]-y[0])**2+(x[1]-y[1])**2+(x[2]-y[2])**2

def least_squares_with_distribution(x,y,dist,eps): 
    # Функционал ошибки учитывающий вероятности данных значений.
    prob=[dist[i].cdf(x[i]+eps[i])-dist[i].cdf(x[i]-eps[i]) for i in range(len(dist))]
    return least_squares(x,y)*(1-prob[0])*(1-prob[1])*(1-prob[2])
```

```python
conditions = ({'type': 'eq', 'fun' : lambda x: np.array([x[0] + x[1] - x[2]])}) #условие x[0]+x[1]-x[2]=0
                                     
for i in range(10):
    y=np.array([norm_in1.rvs(1),norm_in2.rvs(1),norm_out.rvs(1)])
    res1 = sci.optimize.minimize(least_squares, x0=y, args=y, constraints=conditions)
    
    eps=[3,3,3]
    res2 = sci.optimize.minimize(least_squares_with_distribution, 
                                x0=y, args=(y,dist,eps), constraints=conditions)
    print("Измерения:",*y.round(4))
    print("LS values:",res1.x.round(4))
    print("LSwD values:",res2.x.round(4))
    print("distance:",round(np.square(res1.x-res2.x).mean(),5),"delta",(res1.x-res2.x).round(6),"\n")
```

Возможно некоторые из этих распределений даже не стоит менять и лучше оставить такими, какие они есть? В этой библиотеке существует два класса распределений: наблюдаемые и ненаблюдаемые. В нашем случае in1,in2,out не наблюдаются, потому что целиком и полностью зависят от наблюдаемых величин mu_in1, mu_in2 и тд. Одна из идей обхода этой проблемы - заменить в уравнениях связи "не моделируемые" величины на их средние значения.
