### С запасом определяем количество горбов (т.к. начальное приближение осуществляется случайным образом)

```python
k = 50
X = df["Tpp"]
bins = np.linspace(np.min(X),np.max(X),100)
weights = np.ones((k)) / k
means = np.random.choice(X, k)
variances = np.random.random_sample(size=k)
print(means, variances)
```

```python
def pdf(data, mean: float, variance: float):
    # Непрерывная нормально распределенная случайная величина.
    # Расчет её плотности вероятности
    s1 = 1/(np.sqrt(2*np.pi*variance))
    s2 = np.exp(-(np.square(data - mean)/(2*variance)))
    return s1 * s2
```

### Реализация EM  алгритма

```python
# eps нужно для того, чтобы алгоритм не падал на M шаге из-за деления на 0, если будут нулевые вероятности
eps=1e-8
for step in range(25):
  
  if step % 1 == 0:
    plt.figure(figsize=(10,6))
    axes = plt.gca()
    plt.xlabel("$x$")
    plt.ylabel("pdf")
    plt.title("Iteration {}".format(step))
    plt.scatter(X, [0.005] * len(X), color='navy', s=30, marker=2, label="Train data")

    #Промежуточная отрисовка горбов для просмотра динамики (только первых 5)

    plt.plot(bins, pdf(bins, means[0], variances[0]), color='blue', label="Cluster 1")
    plt.plot(bins, pdf(bins, means[1], variances[1]), color='green', label="Cluster 2")
    plt.plot(bins, pdf(bins, means[2], variances[2]), color='magenta', label="Cluster 3")
    plt.plot(bins, pdf(bins, means[3], variances[3]), color='magenta', label="Cluster 4")
    plt.plot(bins, pdf(bins, means[4], variances[4]), color='magenta', label="Cluster 5")
    
    plt.legend(loc='upper left')
    plt.show()
  
    # Для подсчета максимума функции правдоподобия для каждого измерения xi
    likelihood = []
  

    # E шаг
    for j in range(k):
        likelihood.append(pdf(X, means[j], np.sqrt(variances[j])))
    likelihood = np.array(likelihood)
    
    b = []
    # M шаг
    for j in range(k):
        # Используем текущие значения параметров для оценки апостериорной вероятности данных,
        # которые были получены каждой гауссианой
        b.append((likelihood[j] * weights[j]) / (np.sum([likelihood[i] * weights[i] for i in range(k)], axis=0)+eps))

        # Вычисляем матожидания и дисперсии
        # По сути в явном виде условные матожидания и дисперсии
        means[j] = np.sum(b[j] * X) / (np.sum(b[j]+eps))
        variances[j] = np.sum(b[j] * np.square(X - means[j])) / (np.sum(b[j]+eps))

        # Обновляем веса
        weights[j] = np.mean(b[j])
```

### Совмещение одинаковых горбов (с матожиданиями отличающимися менее, чет на 1.5%)

```python
def reduce_num(*args):
    res_m = []
    res_c = []
    res_w = []
    len_m = len(args[0])
   
    for i in range(len_m):
        flag = 0
        pos = 0
        
        for j in range(len(res_m)):
            if(abs(args[0][i]-res_m[j]) < (0.015*args[0][i])):
                flag = 1
                
                pos = j
                
                break
        if flag == 1:

            res_c[pos] = (res_c[pos]+args[1][i])/2
            res_w[pos] = (res_w[pos]+args[2][i])
        else:

            res_m.append(args[0][i])
            res_c.append(args[1][i])
            res_w.append(args[2][i])
    return res_m,res_c,res_w 
    
```
