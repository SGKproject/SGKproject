# Построение модели в PyMC

```python
#импорт необходимых библиотек
import pandas as pd
import numpy as np
import networkx as nx
import pymc3 as pm
import theano.tensor as tt
import matplotlib.pyplot as plt
%matplotlib inline
```

## Загружаем данные

```python
data = pd.read_excel("ActualDataFeb2020.xlsx")
```

```python
data.head()
```

Пытаемся предсказать распределения переменных. Задаем из в виде нормального и равномерного распределения. 

По сути мы пытаемся использовать подход из имеющего ноутбука, а именно сначала задать переменные с какими-то параметрами, затем добавить в модель наблюдения. 

```python
with pm.Model() as model:
    #переменные (истинные значения)
    Gpp = pm.Normal('Gpp', mu=np.mean(data['Gpp']), sd=np.std(data['Gpp']))
    Tpp = pm.Uniform('Tpp', np.min(data['Tpp']),np.max(data['Tpp']))
    
    Go2 = pm.Normal('Go2', mu=np.mean(data['Go2']), sd=np.std(data['Go2']))
    To2 = pm.Normal('To2', mu=np.mean(data['To2']), sd=np.std(data['To2']))
    
    Go4 = pm.Normal('Go4', mu=np.mean(data['Go4']), sd=np.std(data['Go4']))
    To4 = pm.Uniform('To4', np.min(data['T04']),np.max(data['T04']))
    
    Go5 = pm.Normal('Go5', mu=np.mean(data['Go5']), sd=np.std(data['Go5']))
    To5 = pm.Uniform('To5', np.min(data['To5']),np.max(data['To5']))
    
    GoSN2 = pm.Normal('GoSN2', mu=np.mean(data['GoSN2']), sd=np.std(data['GoSN2']))
    GoSN4 = pm.Normal('GoSN4', mu=np.mean(data['GoSN4']), sd=np.std(data['GoSN4']))
    GoSN5 = pm.Normal('GoSN5', mu=np.mean(data['GoSN5']), sd=np.std(data['GoSN5']))
    
    GpSN4 = pm.Uniform('GpSN4', np.min(data['GpSN4']),np.max(data['GpSN4']))
    GpSN5 = pm.Normal('GpSN5', mu=np.mean(data['GpSN5']), sd=np.std(data['GpSN5']))
    
    ToSN4 =  pm.Normal('ToSN4', mu=np.mean(data['ToSn4']), sd=np.std(data['ToSn4']))
    ToSN5 =  pm.Uniform('ToSN5', np.min(data['ToSN5']),np.max(data['ToSN5']))
    
    #уравнения связи
    Gp2 = pm.Deterministic('Gp2', Go2+GoSN2)
    Go11 = pm.Deterministic('Go11', Go4 + GoSN4 - Go2)
    To11 = pm.Deterministic('To11', (Go4*To4 + GoSN4*ToSN4 - Go2*To2)/Go11)
    Gpb4 = pm.Deterministic('Gpb4', Gp2 + Go11)
    Gp4 = pm.Deterministic('Gp4', Gpb4 - GpSN4)
    Go14 = pm.Deterministic('Go14', Gpp + Go5 + GoSN5)
    To14 = pm.Deterministic('To14', (Gpp*Tpp + Go5*To5 + GoSN5*ToSN5)/Go14)
    Gp5 = pm.Deterministic('Gp5', Go14 - GpSN5)
    
    
```

Добавляем измерения в модель. 

# Вопрос: 
это такое предположение, что измерения это такие же случайные переменные, но с мат.ожиданием равным истинным переменным или этот важно правило, которого стоит придерживаться? 

```python
with model:
    Gpp_ = pm.Normal('Gpp_', mu=Gpp, sd=np.std(data['Gpp']),observed=data['Gpp'])
    Tpp_ = pm.Normal('Tpp_', mu=Tpp, sd=np.std(data['Tpp']), observed=data['Tpp'])
    
    Go2_ = pm.Normal('Go2_', mu=Go2, sd=np.std(data['Go2']), observed=data['Go2'])
    To2_ = pm.Normal('To2_', mu=To2, sd=np.std(data['Gpp']), observed=data['To2'])
    
    Go4_ = pm.Normal('Go4_', mu=Go4, sd=np.std(data['Go4']), observed=data['Go4'])
    To4_ = pm.Normal('To4_', mu=To4, sd=np.std(data['T04']), observed=data['T04'])
    
    Go5_ = pm.Normal('Go5_', mu=Go5, sd=np.std(data['Go5']), observed=data['Go5'])
    To5_ = pm.Normal('To5_', mu=To5, sd=np.std(data['To5']), observed=data['To5'])
    
    GoSN2_ = pm.Normal('GoSN2_', mu=GoSN2, sd=np.std(data['GoSN2']), observed=data['GoSN2'])
    GoSN4_ = pm.Normal('GoSN4_', mu=GoSN4, sd=np.std(data['GoSN4']), observed=data['GoSN4'])
    GoSN5_ = pm.Normal('GoSN5_', mu=GoSN5, sd=np.std(data['GoSN5']), observed=data['GoSN5'])
    
    GpSN4_ = pm.Normal('GpSN4_', mu=GpSN4, sd=np.std(data['GpSN4']), observed=data['GpSN4'])
    GpSN5_ = pm.Normal('GpSN5_', mu=GpSN5, sd=np.std(data['GpSN5']), observed=data['GoSN5'])
   
    ToSN4_ = pm.Normal('ToSN4_', mu=ToSN4, sd=np.std(data['ToSn4']), observed=data['ToSn4'])
    ToSN5_ = pm.Normal('ToSN5_', mu=ToSN5, sd=np.std(data['ToSN5']), observed=data['ToSN5'])
    
    #из уравнений связи
    Gp2_ = pm.Normal('Gp2_', mu=Gp2, sd=np.std(data['Gp2']), observed=data['Gp2'])
    Go11_ = pm.Normal('Go11_', mu=Go11, sd=np.std(data['Go11']), observed=data['Go11'])
    To11_ = pm.Normal('To11_', mu=To11, sd=np.std(data['To11']), observed=data['To11'])
    Gpb4_ = pm.Normal('Gpb4_', mu=Gpb4, sd=np.std(data['Gpb4']), observed=data['Gpb4']) 
    Gp4_ = pm.Normal('Gp4_', mu=Gp4, sd=np.std(data['Gp4']), observed=data['Gp4'])
    Go14_ = pm.Normal('Go14_', mu=Go14, sd=np.std(data['Go14']), observed=data['Go14'])
    To14_ = pm.Normal('To14_', mu=To14, sd=np.std(data['To14']), observed=data['To14'])
    Gp5_ = pm.Normal('Gp5_', mu=Gp5, sd=np.std(data['Gp5']), observed=data['Gp5'])

    
```

```python
#поиск точки старта
with model:
    start = pm.find_MAP()
```

```python
start
```

```python
#поиск распределений
with model:
    step = pm.NUTS()
    trace = pm.sample(6000,start=start,step=step,tune=1000,chains=1)
```

```python
res = pm.trace_to_dataframe(trace)
```

```python
res.hist(bins=60,figsize=(18,14))
plt.show()
```
