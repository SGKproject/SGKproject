```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_excel('Kalman_TA9_sorted_light.xlsx')
```

```python
df['Time'] = df['Time'].apply(lambda x: x.replace('.', ''))   

df['Time'] = df['Time'].apply(lambda x: x.replace(':', ''))   

df['Time'] = df['Time'].apply(lambda x: x.replace('-', ''))   

df['Time'] = df['Time'].apply(lambda x: x.replace(' ', ''))   

#data['Time'] = data['Time'].apply(lambda x: x.replace('022020', ''))   
```

```python

df
```

```python
def mask_time(year,month,day,time):
    month_correct = '{:0>2}'.format(month)
    day_correct = '{:0>2}'.format(day)
    time_correct = '{:0>2}'.format(time)
    time_correct = '{:0<6}'.format(time_correct)
    return f'{year}'+month_correct+day_correct+time_correct
```

```python



for col in df.columns[1:3]:
    iterator=0
    print(f"Распределения показателя {col} по месяцам")
    fig,axes=plt.subplots(7,4)
    fig.set_figheight(20)
    fig.set_figwidth(16)
    year=2018
    month=1
    for j in range(2018,2020):
        year=j
        for i in range(1,11):
            month=i
            df_day= df[(df['Time'] <= mask_time(year,month,31,23)) & (df['Time'] >= mask_time(year,month,1,0))] 
          
          
            axes[iterator//4,iterator%4].hist(df_day[col], label= col, bins = 50)
            plt.legend()
            iterator+=1
    plt.show()       
```
