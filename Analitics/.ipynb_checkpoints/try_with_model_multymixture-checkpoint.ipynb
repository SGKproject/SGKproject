```python
import pandas as pd
import numpy as np
import networkx as nx
import pymc3 as pm
import seaborn as sns
import theano
import theano.tensor as tt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_excel("ActualDataFeb2020.xlsx")
```

```python
data
```

```python
with pm.Model() as model:
    #переменные (истинные значения)
    mu = pm.Normal('mu',0.,5.,shape=2)
    sigma = pm.Normal('sigma',1,1,shape=2)
    
    res=pm.NormalMixture('y',w=[0.5,0.5],
                     mu=mu,
                     sd=1,
                     observed=data['Gp2'])
```

```python
with model:
    start = pm.find_MAP()
```

```python
start
```

```python
with model:
    step = pm.NUTS()
    trace = pm.sample(10000,start=start,step=step,tune=2000,chains=2)
```

```python
pm.traceplot(trace)
```

```python
pm.summary(trace)
```
